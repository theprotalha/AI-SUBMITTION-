```python
from langchain_community.document_loaders import PyPDFLoader

print("TASK 3: LOAD PDF DATA\n")

pdf_loader = PyPDFLoader("lecture_notes.pdf")
pdf_docs = pdf_loader.load()

print("Total Pages in PDF:", len(pdf_docs))
print("\nFirst Page Content:\n")
print(pdf_docs[0].page_content)
print("\nPDF Metadata:")
print(pdf_docs[0].metadata)
```

```python
from langchain_community.document_loaders import PyPDFLoader

print("TASK 3: LOAD PDF DATA\n")

pdf_loader = PyPDFLoader("lecture_notes.pdf")
pdf_docs = pdf_loader.load()

print("Total Pages in PDF:", len(pdf_docs))
print("\nFirst Page Content:\n")
print(pdf_docs[0].page_content)
print("\nPDF Metadata:")
print(pdf_docs[0].metadata)
```

```python
from langchain_community.document_loaders import WebBaseLoader

print("TASK 4: LOAD WEB DATA\n")

web_loader = WebBaseLoader(
    "https://en.wikipedia.org/wiki/Retrieval-augmented_generation"
)

web_docs = web_loader.load()

print("Web Page Content (Preview):\n")
print(web_docs[0].page_content[:400])

print("\nWeb Metadata:")
print(web_docs[0].metadata)
```

```python
print("TASK 6: COMPARISON OF DOCUMENT LOADERS\n")

print("PDF Loader:")
print("- Content: Page-wise unstructured text")
print("- Metadata: source file, page number\n")

print("Web Loader:")
print("- Content: Webpage textual content")
print("- Metadata: URL source\n")

print("CSV Loader:")
print("- Content: Row-wise structured data")
print("- Metadata: row index, source file")
```
